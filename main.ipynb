```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('vehicle.csv')
```

```python
df
```

## Scatter Plot Code

```python
fig = plt.figure(figsize = (10,6))
plt.rcParams.update({'font.size': 12})
plt.scatter(df.iloc[2,:].values[2: 16], df.iloc[18,:].values[2: 16], label='Passengers')
plt.scatter(df.iloc[9,:].values[2: 16], df.iloc[18,:].values[2: 16], label='Commercial')
plt.title('Production of different kinds of Vehicles')
plt.legend(title='Type of Vehicle')
plt.xlabel('No. of Passenger/Commerical Vehicles')
plt.ylabel('Total Number of Vehicles')
plt.grid()
plt.savefig('scatter plot.png')
plt.show()
```

## Box Plot Code

```python
years = ['2006-07', '2007-08', '2008-09', '2009-10']
temp_data = df[years][:7]
fig = plt.figure(figsize = (10,6))
plt.rcParams.update({'font.size': 12})

data_list = []
for i in range(4):
    data_list.append(temp_data.iloc[:,i].values)
plt.boxplot(data_list, labels = years, patch_artist = True)
plt.title("Vehicle Production in 4 years")
plt.xlabel('Financial Year')
plt.ylabel('Production')
plt.gca().yaxis.grid(True)
plt.savefig('box plot.png')
plt.show()
```

## Bar Plot Code

```python
years = ['2001-02', '2002-03', '2003-04', '2004-05', '2005-06','2006-07', '2007-08', '2008-09', '2009-10']
temp_data = df[years].iloc[18,:].values

fig = plt.figure(figsize = (10,6))
plt.rcParams.update({'font.size': 12})

plt.bar(years, temp_data)
plt.title("Vehicle Production over years")
plt.xlabel('Financial Year')
plt.ylabel('Total Production')
plt.gca().yaxis.grid(True)
plt.savefig('bar plot.png')
plt.show()
```
